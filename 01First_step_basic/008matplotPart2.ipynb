```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#sine curve

```
np.pi
```

```
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.title("Sine wave form")
plt.plot(x,y)
```

```
x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y_sin,'y--')
plt.title("Sine curve")
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title("Cos curve")
```

#BAr plot

```
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color = 'y')
plt.title('Bar graph')
plt.xlabel('X axis')
plt.ylabel('y axis')
```

#Histogram
density:
x-> all the value
y->count of that value

```
a = np.array([22,887,5,43,56,73,55,54,11,20,51,20,51,5,79,31,27])
plt.hist(a)
plt.title('Histogram')
```

```
a = np.array([22,887,5,43,56,73,55,54,11,20,51,20,51,5,79,31,27])
plt.hist(a,bins =10)
plt.title('Histogram')
```

#box plot
show percentile (0%,25%,...75%)

```
data = [np.random.normal(0,std,5) for std in range(1,4)]
data
```

```
plt.boxplot(data)
```

```
plt.boxplot(data,vert = True,patch_artist=True)
```

```
plt.boxplot(data,vert = False,patch_artist=True)
```
